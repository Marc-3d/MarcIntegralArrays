```julia
using MarcIntegralArrays, Images, VideoIO, ImageView, TestImages
```

```julia
img   = TestImages.testimage( "fabio_color_256" );
scale = (2,2)
L2s   = MarcIntegralArrays.localL2avg( img, (1,1).*2 );
L2avg = MarcIntegralArrays.localAvgs( L2s, (4,4).*2 ); 
L2s   = MarcIntegralArrays.localL2avg( img, (4,4).*2 );
L2s_  = MarcIntegralArrays.localAvgs( L2s, (4,4).*8 ); 
Images.Gray.( L2s .> L2s_)
```
